Тогда сгруппируем их по темам.

```python
def purpose_category(purpose):
    if 'автомобил' in purpose:
        return 'операции с автомобилем'
    elif 'жиль' in purpose or 'недвижимост' in purpose:
        return 'операции с недвижимостью'
    elif 'свадьб' in purpose:
        return 'проведение свадьбы'
    elif 'образовани' in purpose:
        return 'получение образования'
data['purpose_category'] = data['purpose'].apply(purpose_category)
```

Гипотеза: семейное положение человека влияет на возврат кредита в срок.\
Проведем A/B тестирование:\
H0: семейное положение влияет на возврат кредита.\
H1: семейное положение не влияет на возврат кредита\
Уровень значимости равен 0.05.

```python
#chi2-test 

alpha = .05

# Создание таблицы сопряженности(кросс таблицы)
contingency_table = pd.crosstab(data['family_status'], data['debt'])

results = st.chi2_contingency(contingency_table)

if results.pvalue < alpha:
    print('Нулевая гипотеза принимается')
else:
    print('Нулевая гипотеза отклоняется')
```

# Подготовка данных к обучению

```python
x1 = data[['days_employed']]
x2 = data[['total_income']]
y = np.array(data[['debt']])
plt.figure(figsize=(100, 10))
scatter = plt.scatter(x1, x2, c=y, cmap='viridis', edgecolor='k', alpha=0.7)
```

Разделим данные на признаки и целевую переменную(задолженность по кредиту). Кроме того, из признаков удалим категориальные переменные с дублирующими закодированными столбцами.

```python
features = data.drop(['education', 'family_status','debt','purpose'],axis=1)
target = data['debt']
```

```python
features.count(), target.count()
```

Разделим выборки на обучающую, валидационную и тестовую

```python
features_train,features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.4, random_state=12345,stratify=target)
features_test,features_valid, target_test, target_valid = train_test_split(features_valid, target_valid, test_size=0.5, random_state=12345,stratify=target_valid)
```

Далее, нам нужно закодировать категориальные данные и отмасштабировать численные признаки.

```python
features.info()
```

Выделим категориальные и численные признаки.

```python
numeric = [ 'days_employed', 'dob_years','total_income']
category = ['gender', 'income_type', 'purpose_category']
```

Стандартизируем наши численные признаки.

```python
scaler = StandardScaler()
features_train[numeric] = scaler.fit_transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

Закодируем наши признаки для Logreg методом OHE.

```python
features_train_ohe = pd.get_dummies(features_train, drop_first=True).astype('int')
features_valid_ohe = pd.get_dummies(features_valid, drop_first=True).astype('int')
features_test_ohe = pd.get_dummies(features_test, drop_first=True).astype('int')
```

Проверим количество признаков в каждой из выборок.

```python
features_train_ohe.shape, features_valid_ohe.shape, features_test_ohe.shape
```

Количество признаков не совпадает. Тогда найдем те признаки, которых не хватает и добавим их со значением 0

```python
def same_features(features_one,features_second):
    features_one.astype('int')
    for col in features_one.columns.to_list():
        if col not in features_second.columns.to_list():
            features_second[col] = 0
```

Функция готова, применим ее.

```python
features_list = [features_train_ohe, features_valid_ohe, features_test_ohe]
for el in features_list:
    for el_sec in features_list:
        same_features(el,el_sec)
```

```python
features_train_ohe.shape, features_valid_ohe.shape, features_test_ohe.shape
```

Расставим колонки в одинаковом порядке.

```python
features_valid_ohe = features_valid_ohe[features_train_ohe.columns]
features_test_ohe = features_test_ohe[features_train_ohe.columns]
```

```python
features_valid
```

Закодируем наши категориальные признаки методом OE

```python
encoder = OrdinalEncoder()

features_train_oe = features_train.copy()
features_valid_oe = features_valid.copy()
features_test_oe = features_test.copy()

features_train_oe[category] = encoder.fit_transform(features_train_oe[category])
features_valid_oe[category] = encoder.transform(features_valid_oe[category])
features_test_oe[category] = encoder.transform(features_test_oe[category])
```

Данные готовы к обучению.

Посмотрим на баланс классов.

```python
print(target_train.sum()/ target_train.count()*100)
print(target_valid.sum()/ target_valid.count()*100)
target_test.sum()/ target_test.count()*100
```

Как мы видим данные сильно несбаланисрованны, а значит нам это нужно учитывать при обучении модели.

Рассмотрим наши данные на мультиколлинеарность, и исключим сильно скоррелированные признаки.

Удалим признаки, корреляция которых >0.8. Между признаками income_type_пенсионер и days_employed она равна 0.93. Удалим первый признак.

```python
features_train_ohe = features_train_ohe.drop('income_type_пенсионер', axis=1)
features_valid_ohe = features_valid_ohe.drop('income_type_пенсионер', axis=1)
features_test_ohe = features_test_ohe.drop('income_type_пенсионер', axis=1)
```

Теперь сравним значение метрики константной модели с остальными моделями.

# KNN

Подберем гиперпараметры для KNN( количество соседей) для метрики f1_score на валидационной выборке.

```python

#Подберем гиперпараметры на валидационной выборке

f1 = 0
for n in range(1,21):
    model = KNeighborsClassifier(weights='distance',n_neighbors=n)
    model.fit(features_train_oe,target_train)
    y_pred = model.predict_proba(features_valid_oe)[:,1]
    precision, recall, treshold = precision_recall_curve(target_valid, y_pred)
    f1_scores_ar = 2*precision*recall/(precision+recall)
    best_treshold = treshold[np.argmax(f1_scores_ar)]
    pred_best_treshold = (y_pred > best_treshold).astype('int')
    f1_score_new = f1_score(target_valid,pred_best_treshold)
    if f1 < f1_score_new:
        f1 = f1_score_new
        best_n = n
```

```python
print('Наибольшее значение метрики f1_score:', f1,',достигается при количестве соседей равное -', best_n)
```

# LogReg 

Создадим модель логистической регрессии.

```python
model = LogisticRegression(class_weight='balanced', random_state=12345)
```

Обучим на тренировочной выборке.

```python
model.fit(features_train_ohe,target_train)
```

Подберем значение threshold для классификации с максимальным значением метрики f1_score

```python
#Массив вероятностей
y_probs = model.predict_proba(features_valid_ohe)[:,1]
```

```python
#Посчитаем значение precision и recall для порогового значения
precision, recall, thresholds_pr = precision_recall_curve(target_valid, y_probs)
```

```python
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```

```python
#Список значений метрик по threshold
f1_scores_ = 2*precision*recall / (precision+recall)
```

```python
#Значение лучшего порога
best_treshold_pr = thresholds_pr[np.argmax(f1_scores_)]
```

```python
#Предсказания для лучшего порога
pred_best_treshold = (y_probs >= best_treshold_pr).astype('int')
```

```python
#Значение метрики на валидационных данных
f1_score(target_valid, pred_best_treshold)
```

```python
precision, recall, threshold = precision_recall_curve(target_valid, pred_best_treshold)
```

Таким образом, метрика logreg на нашей выборке показала 0.20422898401237755.

# Решающее дерево

Подберем гиперпараметры для DecisionTree для метрики f1_score на валидационной выборке.

```python
f1 = 0
best_depth = 1
for depth in range(1,100):
    model = DecisionTreeClassifier(class_weight='balanced',max_depth=depth, random_state=12345)
    model.fit(features_train_oe,target_train)
    y_pred = model.predict_proba(features_valid_oe)[:,1]
    precision, recall, treshold = precision_recall_curve(target_valid, y_pred)
    f1_scores_ar = 2*precision*recall/(precision+recall)
    best_treshold = treshold[np.argmax(f1_scores_ar)]
    pred_best_treshold = (y_pred > best_treshold).astype('int')
    f1_score_new = f1_score(target_valid,pred_best_treshold)
    if f1 < f1_score_new:
        f1 = f1_score_new
        best_depth = depth
```

```python
print('Наибольшее значение метрики f1_score:', f1,',достигается при глубине равной -', best_depth)
```

# RandomForest

Подберем гиперпараметры для RandomForestна валидационной выборке для метрики f1_score.

```python
%%time

#Подберем гиперпараметры на валидационной выборке

f1 = 0
for n in range(500,1001,100):
    for depth in range(3,7):
        model = RandomForestClassifier(class_weight='balanced',max_depth=depth,n_estimators=n,random_state = 12345)
        model.fit(features_train_oe,target_train)
        y_pred = model.predict_proba(features_valid_oe)[:,1]
        precision, recall, treshold = precision_recall_curve(target_valid, y_pred)
        f1_scores_ar = 2*precision*recall/(precision+recall)
        best_treshold = treshold[np.argmax(f1_scores_ar)]
        pred_best_treshold = (y_pred > best_treshold).astype('int')
        f1_score_new = f1_score(target_valid,pred_best_treshold)
        if f1 < f1_score_new:
            f1 = f1_score_new
            best_n = n
            best_depth = depth
```

```python
print('Наибольшее значение метрики f1_score:', f1,',достигается при количестве деревьев равное -', best_n, ',глубине -', best_depth)
```

# Catboost

Подберем гиперпараметры для CatBoost на валидационной выборке для метрики f1_score.

```python
%%time

#Подберем гиперпараметры на валидационной выборке

f1 = 0
for n in range(500,1001,100):
    for depth in range(2,6):
        model = CatBoostClassifier(scale_pos_weight=9,depth=depth,n_estimators=n,random_state = 12345)
        model.fit(features_train_oe,target_train)
        y_pred = model.predict_proba(features_valid_oe)[:,1]
        precision, recall, treshold = precision_recall_curve(target_valid, y_pred)
        f1_scores_ar = 2*precision*recall/(precision+recall)
        best_treshold = treshold[np.argmax(f1_scores_ar)]
        pred_best_treshold = (y_pred > best_treshold).astype('int')
        f1_score_new = f1_score(target_valid,pred_best_treshold)
        if f1 < f1_score_new:
            f1 = f1_score_new
            best_n = n
            best_depth = depth
```

```python
print('Наибольшее значение метрики f1_score:', f1,',достигается при количестве деревьев равное -', best_n, ',глубине -', best_depth)
```
